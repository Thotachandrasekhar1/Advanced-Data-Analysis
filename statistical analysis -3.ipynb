### What is the Average price of vehicle by fuel type and gearbox type.Give a plot


```python
a=pd.crosstab(index=df["vehicleType"],columns=df["gearbox"],values=df["price"],aggfunc="mean")
```

###  What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this


```python
d=pd.crosstab(index=df["brand"],columns=df["vehicleType"],values=df["price"],aggfunc="mean")
```

```python
d
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(d, annot=True, cmap='YlGnBu', fmt='.0f', cbar_kws={'label': "Average Price"})
plt.title('Average Vehicle Prices by Brand and Type')
plt.show()
```
